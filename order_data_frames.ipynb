```python
import pandas as pd
```

```python
df = pd.read_csv("Top-Movies.csv")
df
```

```python
#Sort values for a specific column by default will be ascending as True
df_ordered = df.sort_values(by="rating")
df_ordered.head(10)
```

```python
#Sort values for a specific column changing ascending as False, descending
df_ordered = df.sort_values(by= "rating", ascending= False)
df_ordered
```
